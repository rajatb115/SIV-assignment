# Reading the dataset

```python
import pandas as pd
import csv

# reading the csv data from file
data = open("Data/malicious_phish.csv")
#dataset = csv.reader(data)
dataset = pd.read_csv(data)
print("Type :",type(dataset))
print("Dataset shape :",dataset.shape)


# reading the headers of the data
header = []
#header = next(dataset)
header = dataset.columns
print("Header :",header)

url_cnt = 0
url_data = []


#print(dataset['type'][651191])
for i in tqdm(range(dataset.shape[0]),desc="Processing data"):
    
    val = 0
    
    if dataset['type'][i]=="benign":
        val = -1
    elif dataset['type'][i] == "other":
        continue
    else:
        val = 1
    
    url_data.append([dataset['url'][i],val])

url_cnt = len(url_data)
```
